```python
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent))
from mnist import get_mnist_dataset

import holoviews as hv
hv.extension("bokeh")

hv.plotting.bokeh.ElementPlot.height = 500
hv.plotting.bokeh.ElementPlot.width = 500
```

```python
import numpy as np
np.random.seed(5090)
```

```python
X_train, Y_train, X_test, Y_test = get_mnist_dataset()
```

```python
def with_dropout():
    from mini_dl.nn.model import Sequential
    from mini_dl.nn.layer import Dense, ReLU, Dropout
    from mini_dl.optim.optimizer import SGDOptimizer, Adam
    from mini_dl.nn.loss import CrossEntropy
    
    model = Sequential([
        Dense(784, 250),
        ReLU(),
        Dropout(),
        Dense(250, 250),
        ReLU(),
        Dropout(),
        Dense(250, 10),
    ])

    model.compile_(optimizer=SGDOptimizer(momentum=0.9, nesterov=True), loss=CrossEntropy())

    hist = model.fit(X_train, Y_train, epochs=100, batch_size=128, validation_split=0.3)
    return hist
```

```python
def no_dropout():
    from mini_dl.nn.model import Sequential
    from mini_dl.nn.layer import Dense, ReLU
    from mini_dl.optim.optimizer import SGDOptimizer, Adam
    from mini_dl.nn.loss import CrossEntropy
    
    model = Sequential([
        Dense(784, 250),
        ReLU(),
        Dense(250, 250),
        ReLU(),
        Dense(250, 10),
    ])

    model.compile_(optimizer=SGDOptimizer(momentum=0.9, nesterov=True), loss=CrossEntropy())

    hist = model.fit(X_train, Y_train, epochs=100, batch_size=128, validation_split=0.3)
    return hist
```

```python
dropout_hist = with_dropout()
nodrop_hist = no_dropout()
```

```python
hv.Overlay([hv.Curve(h) for h in dropout_hist.values()]).opts(title="Dropout") + \
hv.Overlay([hv.Curve(h) for h in nodrop_hist.values()]).opts(title="No Dropout")
```
